# Abstract

**Objective**: Explore the most common commands to edit graphs using the `Axes` object. It can be useful for example to edit plots done with `pandas`.

# Initialize

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

## Default parameters

```python
print(mpl.rcParamsDefault.keys())

# Default parameters in:
# mpl.rcParamsDefault
# mpl.rcParams
```

# Explore commands

```python
# create figure and axes objects
# =============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

# alternative
# fig = plt.figure(figsize=(4,4))
# ax = fig.add_subplot(111)

# edit after creation
# -------------------------------------
# fig.set_size_inches(10, 10)


# plot something
# =============================================================================
ax.plot([-10,-5,5,10], [1,3,5,10], label='line1')
pl2, = ax.plot([-10,-5,5,10], [2,4,6,10], label='line3')
pl2.set_label('line2')  # for the legend


# figure title
# =============================================================================
fig.suptitle('this is the figure title', fontsize=12, y=1.05)


# axes title
# =============================================================================
ax.set_title('this is the title', fontsize=32)
# text = ax.title.get_text()

# edit after creation
# -------------------------------------
ax.title.set_fontsize(12)


# grid
# =============================================================================
ax.grid(b=True, axis='x', which='major')


# axes
# =============================================================================
ax.set_xlabel('this is the x-label',
              fontsize=32,
              rotation=45)
ax.set_ylabel('this is the y-label',
              fontsize=32,
              labelpad=-10)
# text = ax.xaxis.label.get_text()

# edit after creation
# -------------------------------------
ax.xaxis.label.set_fontsize(12)
ax.xaxis.label.set_rotation(30)
ax.set_xlim(-20, 20)
ax.yaxis.labelpad = 20


# ticks
# =============================================================================

ax.tick_params(axis='both', labelsize=20, colors='blue')

# alternative:
ax.xaxis.set_tick_params(labelsize=12, colors='red')

# edit after creation
# -------------------------------------
for label in ax.get_xticklabels():
    label.set_rotation(-30)
    label.set_fontsize(16)
    # alternative:
    # label.update({'rotation': -30, 'fontsize': 16})

# scientific notation
# -------------------------------------
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useOffset=50)
fmt = ax.xaxis.get_major_formatter() # or: ax.xaxis.major.formatter
fmt._useMathText = True  # e.g., 5x10^1 instead of 5e1

# alternative:
# xfmt = mpl.ticker.ScalarFormatter(useMathText=True, useOffset=50)
# xfmt.set_powerlimits((0,0))
# ax.xaxis.set_major_formatter(xfmt)

# alternative:
# fmt.set_powerlimits((0, 0))
# fmt.set_useOffset(50)

# edit offset text
ax.xaxis.offsetText.set_fontsize(14)
ax.xaxis.offsetText.set_color('green')
# ax.xaxis.get_offset_text().set_fontsize(14) # same as above


# legend
# =============================================================================

# first label, second label, ... (in this order):
# ax.legend(['this is the legend'], fontsize=16, loc='lower right')

# choose the label:
# ax.legend((pl2,), ('this is the legend',), fontsize=16, loc='lower right')

# all labels:
ax.legend(fontsize=16, loc='lower right', bbox_to_anchor=(1.55,0.7))

fontprop = {'size': 24}  # same as: mpl.font_manager.FontProperties(size=18)
ax.legend_.set_title('this is the legend title', prop=fontprop)

# edit after creation
# -------------------------------------
ax.legend_.prop.set_size(12)
# ax.legend_.set_visible(False) # hide legend


# show
# =============================================================================
# fig.show()
# plt.show()
```
